```python
def convert_date(bytes_input):
    str_input = bytes_input.decode("utf-8")
    month, day, year = str_input.split("/")
    if int(year) < 50:
        new_year = "20" + year
    else:
        new_year = "19" + year
    return "/".join([month, day, new_year])
```
